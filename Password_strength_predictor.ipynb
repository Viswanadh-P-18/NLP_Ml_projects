```python
def charsplit(word):
    character=[]
    for i in word:
        character.append(i)
    return character
```

```python
charsplit("sundayanakura")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=charsplit)
```

```python
X=vectorizer.fit_transform(x)
```

```python
features=vectorizer.get_feature_names()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
x_train.shape
```

# using multinomial logistic regression as there are 3 labels

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(multi_class='multinomial')
```

```python
ml=lr.fit(x_train,y_train)
```

```python
predictions=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,predictions)
sns.heatmap(cm,annot=True,cmap='coolwarm')
```

# get the characters with the highest and least weights(coefficients)...

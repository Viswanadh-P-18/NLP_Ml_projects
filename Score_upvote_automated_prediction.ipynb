```python
def textfit(X,y,ml_model,nlp_model,coef_show=1):
    X_c=nlp_model.fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X_c,y)
    ml=ml_model.fit(x_train,y_train)
    acc=ml_model.score(x_test,y_test)
    print(acc)
    
    if coef_show==1:
        w=c.get_feature_names()
        coef=ml.coef_.tolist()[0]
        coef_df=pd.DataFrame({'Coeff':coef,'Words':w})
        coef_df=coef_df.sort_values(['Coeff','Words'],ascending=False)
        print("The top 20 words :")
        print("\n")
        print(coef_df.head(20))
        print("The below 20 words :")
        print("\n")
        print(coef_df.tail(20))
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
def pre(X,y,nlp_model,ml_model):
    X_c=nlp_model.fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X_c,y)
    ml_model.fit(x_train,y_train)
    pred=ml_model.predict(x_test)
    print(x_test.shape)
    print(accuracy_score(y_test,pred))
    cm=confusion_matrix(y_test,pred)
    sns.heatmap(cm,annot=True,cmap='coolwarm')
```

# neglected 40 to 60 as it is neutral

```python
y_dict={'80-100%':1,'60-80%':1,'20-40%':0,'0-20%':0}
```

```python
tf2=TfidfVectorizer()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
def predupvote(X,y,nlp_model,ml_model):
    X_tf=nlp_model.fit_transform(X)
    x_train,x_test,y_train,y_test=train_test_split(X_tf,y)
    ml_model.fit(x_train,y_train)
    p=ml_model.predict(x_test)
    print("The accuracy is",accuracy_score(p,y_test))
    cm=confusion_matrix(p,y_test)
    sns.heatmap(cm,annot=True)
    
    
```

```python
predupvote(X,y,tf2,lr)
```

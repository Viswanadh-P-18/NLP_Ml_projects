```python
def char_to_list(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
        
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer(tokenizer=char_to_list)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(multi_class='multinomial')
```

```python
predictions=lr.predict(x_test)
```

```python
def predicts(features):
    res=[]
    res.append(features)
    a=tf.transform(res)
    d=lr.predict(a)
    if(d[0]==0):
        print("The password is not strong")
    elif(d[0]==1):
        print("The password is strong")
    else:
        print("The password s very strong")
    
```
